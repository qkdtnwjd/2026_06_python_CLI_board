```python
#파이썬 심화

print("== CLI 게시판 실행==")

i=1
titles=[]
contents=[]

while True:

  a = input("명령어)")

  if a == "exit":
    ("== CLI 게시판 종료==")
    break

  if a == "article write":


    title = input("제목)")

    c = input("내용)")

    titles.append(title)
    contents.append(c)

    print("제목:{}".format(title))
    print("내용:{}".format(c))
    print("{}번 글이 생성 되었습니다.".format(i))

    # print(titles)

    i+=1

  if a == "article list":


    print("======================")
    print("번호  /  제목  /  내용 ")

    for num in range(len(titles)):
      print(" {}   /  {}  /   {}".format(num+1,titles[num],contents[num]))

    print("=======================")

    continue


```
